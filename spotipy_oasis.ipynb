```python
# Autenticação com o Spotify
auth_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(auth_manager=auth_manager)

# Buscar o artista Oasis pelo nome
results = sp.search(q='artist:Oasis', type='artist')
oasis_id = results['artists']['items'][0]['id']

# Buscar as 10 músicas mais populares do Oasis
tracks = sp.artist_top_tracks(oasis_id)

# Lista para armazenar dados das faixas
track_data = []

# Mapeamento de tonalidades de números para nomes
key_mapping = {
    0: 'C', 1: 'C♯/D♭', 2: 'D', 3: 'D♯/E♭', 4: 'E', 5: 'F',
    6: 'F♯/G♭', 7: 'G', 8: 'G♯/A♭', 9: 'A', 10: 'A♯/B♭', 11: 'B'
}

# Iterar sobre as faixas para coletar as métricas
for track in tracks['tracks']:
    track_id = track['id']
    track_name = track['name']
    popularity = track['popularity']
    album_name = track['album']['name']
    release_date = track['album']['release_date']
    release_year = release_date.split('-')[0]  # Extrair apenas o ano

    # Buscar características de áudio da faixa
    audio_features = sp.audio_features(track_id)[0]

    # Converter a tonalidade de número para nome
    key_name = key_mapping.get(audio_features['key'], 'Desconhecido')

    # Adicionar os dados da faixa à lista
    track_data.append({
        'Nome da Faixa': track_name,
        'Popularidade': popularity,
        'Nome do Disco': album_name,
        'Ano de Lançamento': release_year,
        'Dançabilidade': audio_features['danceability'],
        'Energia': audio_features['energy'],
        'Valência': audio_features['valence'],
        'Acústica': audio_features['acousticness'],
        'Duração (ms)': audio_features['duration_ms'],
        'Tonalidade': key_name,  # Nome da tonalidade
        'Modo': 'Maior' if audio_features['mode'] == 1 else 'Menor',  # 1 = maior, 0 = menor
        'Tempo (BPM)': audio_features['tempo']  # Tempo da faixa em BPM
    })

# Criar DataFrame com os dados coletados
df = pd.DataFrame(track_data)

# Exibir o DataFrame
print(df)
```
